```python
import pandas as pd
import numpy as np

# Criando o DataFrame
dados = {
    "Data": ["15/01/2025"] * 5,
    "Cidade": ["São Paulo", "Rio de Janeiro", "Curitiba", "Porto Alegre", "Salvador"],
    "Temperatura Máxima (°C)": [30.5, 35.0, 24.0, 28.0, 31.0],
    "Temperatura Mínima (°C)": [22.0, 25.0, 18.0, 20.0, 24.5],
    "Precipitação (mm)": [12.0, np.nan, 8.0, 15.0, np.nan],
    "Umidade Relativa (%)": [78, 70, np.nan, 82, 80]
}
df = pd.DataFrame(dados)

# Tratamento de valores ausentes
df["Precipitação (mm)"] = df["Precipitação (mm)"].fillna(df["Precipitação (mm)"].mean())
df["Umidade Relativa (%)"] = df["Umidade Relativa (%)"].fillna(df["Umidade Relativa (%)"].median())

# Criando a coluna Amplitude Térmica
df["Amplitude Térmica"] = df["Temperatura Máxima (°C)"] - df["Temperatura Mínima (°C)"]

# Criando um DataFrame apenas com cidades com temperatura máxima acima de 30°C
df_filtro = df[df["Temperatura Máxima (°C)"] > 30]

# Reorganizando as colunas
df = df[["Data", "Cidade", "Temperatura Máxima (°C)", "Temperatura Mínima (°C)", "Amplitude Térmica", "Precipitação (mm)", "Umidade Relativa (%)"]]

# Exibindo o DataFrame final
display(df)
```

```python
import matplotlib.pyplot as plt

# Criando os dados do gráfico
horas = list(range(25))
temperaturas = [15 + (i * 1.25) if i <= 12 else 30 - ((i - 12) * 1.0) for i in horas]

# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(horas, temperaturas, marker='o', linestyle='-', color='b', label='Temperatura (°C)')
plt.xlabel("Hora do dia")
plt.ylabel("Temperatura (°C)")
plt.title("Evolução da Temperatura ao Longo do Dia")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Criando os dados
dados_vendas = {
    "Dia": ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"],
    "Vendas Totais": [2000, 2500, 1800, 2200, 2700, 3000, 2800],
    "Clientes": [150, 180, 130, 160, 190, 210, 200],
    "Lucro": [500, 600, 400, 550, 700, 800, 750]
}
df_vendas = pd.DataFrame(dados_vendas)

# Gráfico de barras - Total de vendas por dia
plt.figure(figsize=(8, 5))
sns.barplot(x="Dia", y="Vendas Totais", data=df_vendas, palette="viridis")
plt.title("Total de Vendas por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Vendas Totais")
plt.show()

# Gráfico de dispersão - Clientes x Vendas Totais
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Clientes", y="Vendas Totais", data=df_vendas, color="r")
plt.title("Relação entre Número de Clientes e Vendas Totais")
plt.xlabel("Número de Clientes")
plt.ylabel("Vendas Totais")
plt.show()

# Heatmap - Correlação entre Vendas, Clientes e Lucro
plt.figure(figsize=(6, 4))

# Pegamos apenas colunas numéricas para evitar erro
df_corr = df_vendas.drop(columns=["Dia"])

sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Vendas, Clientes e Lucro")
plt.show()
```
